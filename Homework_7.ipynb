```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Untuk Google Colab :
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
file = '/content/drive/MyDrive/GoogleColab/BankChurners.csv'
df = pd.read_csv(file)

# Tampilkan kolom yang relevan
df = df[['Attrition_Flag', 'Customer_Age', 'Gender', 'Education_Level', 'Income_Category']]
df.head()
```

```python
#1.a. Pie chart
df['Attrition_Flag'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'salmon'])
plt.title('Attrition Flag Distribution')
plt.ylabel('')
plt.show()
```

```python
#1.b. Donut chart
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'pink'])
plt.gca().add_artist(plt.Circle((0, 0), 0.7, color='white'))  # Donut effect
plt.title('Gender Distribution')
plt.show()
```

```python
#1.c. Vertical bar chart
df['Customer_Age'].value_counts().sort_index().plot(kind='bar', color='purple')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#1.d. Horizontal bar chart
df['Education_Level'].value_counts().plot(kind='barh', color='orange')
plt.title('Education Level Distribution')
plt.xlabel('Count')
plt.ylabel('Education Level')
plt.show()
```

```python
#1.e. Bar chart for Income_Category
df['Income_Category'].value_counts().plot(kind='bar', color='green')
plt.title('Income Category Distribution')
plt.xlabel('Income Category')
plt.ylabel('Count')
plt.show()
```

```python
#Pie Chart (Attrition_Flag): Menunjukkan bahwa mayoritas pelanggan tetap (Attrited = rendah).
#Donut Chart (Gender): Distribusi gender hampir merata dengan sedikit lebih banyak perempuan.
#Vertical Bar Chart (Customer_Age): Sebagian besar pelanggan berada di rentang usia menengah (40-60).
#Horizontal Bar Chart (Education_Level): Tingkat pendidikan terbanyak adalah Graduate.
#Bar Chart (Income_Category): Pendapatan kategori "Unknown" mendominasi, menunjukkan data pendapatan pelanggan banyak yang tidak diisi.
```

```python
#3. Countplot memadukan Gender dan Attrition_Flag
sns.countplot(data=df, x='Gender', hue='Attrition_Flag', palette='Set2')
plt.title('Gender vs Attrition Flag')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#4. Scatterplot Customer Age vs Income Category
sns.scatterplot(data=df, x='Customer_Age', y='Income_Category', hue='Attrition_Flag', palette='coolwarm')
plt.title('Customer Age vs Income Category')
plt.xlabel('Customer Age')
plt.ylabel('Income Category')
plt.show()
```

```python
#5. Heatmap korelasi
# Pertama, pastikan data numerik saja untuk heatmap
df_corr = df[['Customer_Age']].copy()  # Tambah kolom numerik lain jika ada
correlation_matrix = df_corr.corr()

# Visualisasi heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='Blues')
plt.title('Heatmap of Numerical Columns')
plt.show()
```

```python
#Countplot:

#Gender vs Attrition_Flag: Pelanggan perempuan lebih cenderung bertahan dibanding laki-laki.

#Scatterplot:

#Customer_Age vs Income_Category: Tidak ada pola langsung antara usia dan pendapatan pelanggan.

#Heatmap:

#Correlation Heatmap: Korelasi antara variabel numerik rendah, menunjukkan ketergantungan antar kolom numerik minimal.
```
